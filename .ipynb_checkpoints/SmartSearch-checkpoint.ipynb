```python
import os

def encontrar(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)
        else:
            return False

os.chdir('/home/gida3pc/MEGAsync/Admin/Certificado/')
os.getcwd()  

directorio = '/home/gida3pc/MEGAsync/Admin/Certificado/'
listacarpeta = [ name for name in os.listdir(directorio) if os.path.isdir(os.path.join(directorio, name)) ]

listaemails = []

for carpeta in range(len(listacarpeta)):
    if encontrar('listaalumnos','/home/gida3pc/MEGAsync/Admin/Certificado/'+listacarpeta[carpeta]) != 0:
        ruta = encontrar('listaalumnos','/home/gida3pc/MEGAsync/Admin/Certificado/'+listacarpeta[carpeta])
        texto = open(ruta).read().splitlines()
        listaparticipantes=texto[3:]
        for participante in range(len(listaparticipantes)):
            datospersona = listaparticipantes[participante].split(',')
            print datospersona[1]
            listaemails.append(datospersona[1])
```

```python
listaemails = [email.replace(' ', '') for email in listaemails]
listaemails
```

```python
text_file = open("Output.txt", "w")
for item in listaemails:
    text_file.write(item + '\n')
text_file.close()
```

```python
text_file = open("/home/gida3pc/MEGAsync/AmazonSES/Campaign61-80/contactos.txt", "a")
for item in listaemails:
    text_file.write(item + '\n')
text_file.close()
```
